## Ejemplo 3: Resolución de ecuaciones algebraicas

La EDO 

$$(y^5 - 6 y^4 - 5 y^3 - 10 y^2 - 36 y + 56)dx+(y^2+1)dy = 0$$

es autónoma. Encontrar los puntos estacionarios de la ecuación.

```python
y = sp.Symbol("y")
sp.solve(y**5 - 6 * y**4 - 5 * y**3 - 10 * y**2 - 36 * y + 56, y)
```

```python
X,Y=np.meshgrid(np.linspace(-3,8,12), np.linspace(-3,8,12))

U = 1.0
V = (-(Y**5 - 6 * Y**4 - 5 * Y**3 - 10 * Y**2 - 36 * Y + 56)) / (Y**2 + 1)

N = np.sqrt(U ** 2 + V ** 2)#Normalizamos los vectores
U = U / N
V = V / N

plt.quiver(X, Y, U, V) #graficamos los vectores con origen en (x,y) y término (u,v)
plt.xlim([-4, 9]) #limites en eje x
plt.ylim([-4, 9]) #limites en eje y

plt.xlabel('x') #etiqueta eje x
plt.ylabel('y') #etiqueta eje y
plt.grid()# para establecer red en la región
```

## Ejemplo 5: Declaración/Resolución/Comprobación de EDO

Declarar y resolver la ecuación diferencial de Bernoulli usando comando `dsolve`

$$\dfrac{dy}{dx} - y = 2e^x y^2 $$

```python
# Se declara como variable la EDO
y = sp.Function("y")
x = sp.Symbol("x")
EDO = sp.Eq(sp.diff(y(x), x), 2*sp.exp(x)*y(x)**2)

# Solución a la EDO
sol = sp.dsolve(EDO, y(x))

# Checkear si la solución es realmente solución de la EDO
sp.checkodesol(EDO, sol)
```

## Ejemplo 7: Calculo de derivadas

Verificar si la EDO es exacta 

$$\left[ \dfrac{\ln(\ln(y))}{x}+\dfrac{2}{3}xy^3+6x \right] dx+ \left[\dfrac{\ln(x)}{y\ln(y)}+x^2y^2+4e^{-2y} \right]dy=0$$


```python
# Se declara M y N, acá se utiliza y como símbolo
y = sp.Symbol("y")
M = sp.log(sp.log(y))/x + (2*x*y**3)/3 + 6*x
N = (sp.log(x))/(y*sp.log(y)) + x**2*y**2 + 4*sp.exp(-2 * y)
# Derivadas parciales
My = sp.diff(M, y)
Nx = sp.diff(N, x)
if My == Nx:
    print("Es exacta")

```

```python
import numpy as np

def bisection(func, a, b, tol=1e-6, max_iter=100):
    """
    Encuentra la raíz de la función 'func' en el intervalo [a, b] utilizando el método de bisección.

    Parámetros:
    - func: La función para la cual encontrar la raíz.
    - a, b: Los extremos del intervalo [a, b].
    - tol: Tolerancia, el algoritmo se detendrá cuando el tamaño del intervalo sea menor que 'tol'.
    - max_iter: Número máximo de iteraciones.

    Devuelve:
    - root: La aproximación de la raíz.
    - iterations: El número de iteraciones realizadas.
    """

    # Verificar si la raíz está en el intervalo [a, b]
    if np.sign(func(a)) == np.sign(func(b)):
        raise ValueError("La función tiene el mismo signo en los extremos del intervalo. No se puede aplicar bisección.")

    # Inicializar variables
    iteration = 0

    while (b - a) / 2 > tol and iteration < max_iter:
        c = (a + b) / 2  # Punto medio del intervalo

        # Verificar si c es una raíz o si se puede acortar el intervalo
        if func(c) == 0:
            root = c
            break
        elif np.sign(func(c)) == np.sign(func(a)):
            a = c
        else:
            b = c

        iteration += 1

    root = (a + b) / 2  # La aproximación final de la raíz

    return root, iteration

# Ejemplo de uso:
# Definir la función, por ejemplo, f(x) = x^2 - 4
def func(x):
    return (x - 4)**2 - 4

# Encontrar la raíz en el intervalo [0, 4]
root, iterations = bisection(func, 5, 9)

print(f"Aproximación de la raíz: {root}")
print(f"Número de iteraciones: {iterations}")
```
