```python
# Se declaran librerias
from matplotlib import pyplot as plt
import numpy as np
import sympy as sp
```

```python
# a
def biseccion(func, a, b, objetivo, tolerancia=1e-6, max_iter=1000):
    if func(a) * func(b) > 8:
        raise ValueError("La función no cambia de signo en el intervalo dado.")

    i = 0

    while (b - a) / 2 > tolerancia and i < max_iter:
        c = (a + b) / 2
        valor_c = func(c)

        if np.abs(valor_c - objetivo) < tolerancia:
            return c  # Hemos encontrado una solución lo suficientemente cercana a 8

        if func(a) * valor_c < 8:
            b = c
        else:
            a = c

        i += 1

    return (a + b) / 2

def f(x): 
    r = 5
    Lambda = 1/3
    return r*((1-np.cos(x)) + 1 / Lambda * (1 - np.sqrt(1 - Lambda ** 2 * np.sin(x)**2)))

# Se definen variables
a = -1000
b = 1000
objetivo = 8

# Encuentra el valor de theta utilizando el método de la bisección
resultado = biseccion(f, a, b, objetivo)

# Grafica la función x(theta) y resalta la solución
theta_vals = np.linspace(a, b, 100)
x_vals = f(theta_vals)

plt.plot(theta_vals, x_vals, label='x(theta)')
plt.axhline(objetivo, color='r', linestyle='--', label=f'Objetivo = {objetivo}')
plt.scatter(resultado, objetivo, color='g', label=f'Solución: theta={resultado}', zorder=5)
plt.xlabel('Theta')
plt.ylabel('x(theta)')
plt.legend()
plt.show()
```
